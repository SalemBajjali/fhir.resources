```python
import json

def convert_to_json(pydantic_obj):
    # Convert to dictionary
    pydantic_dict = pydantic_obj.dict()

    # Convert dictionary to JSON string
    json_str = json.dumps(pydantic_dict, indent=4)
    
    print(json_str) 
```
